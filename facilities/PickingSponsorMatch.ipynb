```python
def string_min(s):
    '''
    minifies a string for comparison purposes
    s: a string
    '''
    return ''.join(w for w in s.lower() if w not in set(string.punctuation + ' '))
```

```python
perfect = [k for k, v in sponsor_possibilities.items() 
           if len(v) > 0 and 
              string_min(v[0]['name']) == string_min(k.decode('utf-8'))]
```
