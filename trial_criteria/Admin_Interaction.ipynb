```python
def human_checker_initial(concept_name, terms, term_set):
    '''This function loops through the possible terms for a concept and
    lets admin input decide if they actually are right for the concept or not'''
    
    print '''NOTE: you must go through all terms in a concept before they can be written
    to the production tables \n'''
    print 'Do the following terms belong in this concept: %r?' % (concept_name)
    if len(terms) > 1:
        for term in terms:
            print 'Term: \x1b[35m %s \x1b[0m' % (term)
            answer_switch = True
            while answer_switch:
                add_term = raw_input('Does this term belong in the %s concept? (Y, N, exit): ' % (concept_name))
                if add_term.lower() == 'y':
                    term_set.update([term])
                    answer_switch = False
                elif add_term.lower() == 'exit':
                    #pass switch to exit program
                    exit_switch = True
                    return term_set, exit_switch
                elif add_term.lower() == 'n':
                    answer_switch = False
                else:
                    pass

    exit_switch = False
    return term_set, exit_switch
```

```python
def review_concepts(concept_id, engine):
    exit_switch = False
    term_set = set()
    
    #check if the concept is new or existing
    new_concept = engine.execute(select([CriteriaConceptStaging.c.new_concept], distinct=True).\
                                  where(CriteriaConceptStaging.c.concept_id ==
                                       concept_id))
    new_concept = [r[0] for r in new_concept][0]
    
    concept_name = engine.execute(select([CriteriaConceptStaging.c.value]).\
                                  where(CriteriaConceptStaging.c.concept_id ==
                                       concept_id).\
                                  where(CriteriaConceptStaging.c.update_type ==
                                       'concept-name'))
    concept_terms = engine.execute(select([CriteriaConceptStaging.c.value]).\
                                   where(CriteriaConceptStaging.c.concept_id ==
                                       concept_id).\
                                   where(CriteriaConceptStaging.c.update_type ==
                                       'term'))
    concept_name = [r[0] for r in concept_name][0]
    concept_terms = [r[0] for r in concept_terms]
    
    
    if new_concept == 1:
        #concept is new
        print 'This is a new concept: %s \n' % (concept_name)
        print 'Here are the terms: \n'
        print concept_terms
        answer_switch = True
        while answer_switch:
            create_concept = raw_input('''Could all or some of the above terms make up a sufficient group
         for the %s concept? (Y, N, exit)''' % (concept_name))
            if create_concept.lower() == 'y':
                choice = True
                answer_switch = False
            elif create_concept.lower() == 'exit':
                #pass switch to exit program
                exit_switch = True
                choice = False
                return exit_switch, choice, term_set, concept_name
            elif create_concept.lower() == 'n':
                choice = False
                return exit_switch, choice, term_set, concept_name
            else:
                pass
        term_set, exit_switch = human_checker_initial(concept_name, concept_terms, term_set)
        
    else:
        #concept is not new, need to pull old concept data to compare
        existing_concept_terms = engine.execute(select([ConceptTerms.c.term]).\
                                   where(ConceptTerms.c.concept_id ==
                                       concept_id))
        existing_concept_terms = [r[0] for r in existing_concept_terms]
        
        #get new terms in the staging table
        new_concept_terms = set(concept_terms).difference(set(existing_concept_terms))
        print 'This is an existing concept. \n'
        print 'Here are the existing terms in the %s concept: \n' % (concept_name)
        print existing_concept_terms
        
        print 'Should the following terms be added to the %s concept? \n' % (concept_name)
        print new_concept_terms
        answer_switch = True
        while answer_switch:
            create_concept = raw_input('Y, N, exit')
            if create_concept.lower() == 'y':
                choice = True
                answer_switch = False
            elif create_concept.lower() == 'exit':
                #pass switch to exit program
                exit_switch = True
                choice = False
                return exit_switch, choice, term_set, concept_name
            elif create_concept.lower() == 'n':
                choice = False
                return exit_switch, choice, term_set, concept_name
            else:
                pass
        term_set, exit_switch = human_checker_initial(concept_name, new_concept_terms, term_set)
    
    
    
    print concept_name, concept_terms
    
    return exit_switch, choice, term_set, concept_name
```

```python
def transfer_concept(concept_id, engine, term_set, concept_name):
    #pull other data for concept
    concept_terms_reject = [r[0] for r in engine.execute(select([CriteriaConceptStaging.c.value]).\
                                   where(CriteriaConceptStaging.c.concept_id ==
                                       concept_id).\
                                   where(CriteriaConceptStaging.c.update_type ==
                                       'term-reject'))]
    concept_predictors_reject = [r[0] for r in engine.execute(select([CriteriaConceptStaging.c.value]).\
                                   where(CriteriaConceptStaging.c.concept_id ==
                                       concept_id).\
                                   where(CriteriaConceptStaging.c.update_type ==
                                       'predictor-reject'))]
    concept_predictors = [r[0] for r in engine.execute(select([CriteriaConceptStaging.c.value]).\
                                   where(CriteriaConceptStaging.c.concept_id ==
                                       concept_id).\
                                   where(CriteriaConceptStaging.c.update_type ==
                                       'predictor'))]
    #write all data into the new tables
    #instert data into db
    engine.execute(CriteriaConcept.insert(), [{
                                      'concept_id': concept_id,
                                      'concept_name':concept_name}])
    engine.execute(ConceptTerms.insert(), [{
                                      'concept_id': concept_id,
                                      'term':term}
                                           for term in term_set])
    engine.execute(ConceptTermsReject.insert(), [{
                                      'concept_id': concept_id,
                                      'term':term}
                                           for term in concept_terms_reject])
    engine.execute(ConceptPredictors.insert(), [{
                                      'concept_id': concept_id,
                                      'predictor':predictor}
                                           for predictor in concept_predictors])
    engine.execute(ConceptPredictorsReject.insert(), [{
                                      'concept_id': concept_id,
                                      'predictor':predictor}
                                           for predictor in concept_predictors_reject])
    
    
    
```

```python
def tag_criteria_sentences(concept_id, engine):
    tagged_sentences = []
    #pull the accepted terms with term_ids
    term_list = engine.execute(select([ConceptTerms.c.term, ConceptTerms.c.term_id]).\
                   where(ConceptTerms.c.concept_id == concept_id)).fetchall()
    
    criteria_text = engine.execute(select([CriteriaText.c.criteria_text,
                                       CriteriaText.c.criteria_text_id]).\
                               where(or_(CriteriaText.c.criteria_text.op('rlike')('[[:<:]]' + t[0] + '[[:>:]]')
                                         for t in term_list))).fetchall()
    
    #find all the sentences where a concept term occurs
    for sent, sent_id in criteria_text:
        for term, term_id in term_list:
            if term in sent.lower():
                tagged_sentences.append((sent_id, term_id))
                
    #write tagged sentences to db
    engine.execute(CriteriaConceptLookup.insert(), [{'criteria_text_id': v[0],
                                                     'term_id': v[1]}
                                                    for v in tagged_sentences])
```

```python
def main():
    '''This gets activated when an admin chooses to review staged criteria concepts'''
    #initialize the connection to the db
    engine = create_engine('mysql+pymysql://' + db_connect.conn)
    metadata.create_all(engine)
    
    #pull all the concept id's of staged concepts
    concept_ids = engine.execute(select([CriteriaConceptStaging.c.concept_id], distinct=True))
    concept_ids = [r[0] for r in concept_ids]
    print concept_ids
    
    print 'Would you like to start reviewing concepts?'
    answer_switch = True
    while answer_switch:
        start = raw_input('Y, N: ')
        if start.lower() == 'y':
            start_choice = True
            answer_switch = False
        elif start.lower() == 'n':
            start_choice = False
            answer_switch = False
        else:
            pass
    if start_choice:
        for concept in concept_ids:
            exit_switch, choice, term_set, concept_name = review_concepts(concept, engine)
            if exit_switch:
                break
            if choice:
                tansfer_concept(concept, engine, term_set, concept_name)
                #transfer data to user history table
                update_user_history()
            else:
                #transfer data to user history table
                update_user_history()
                
```

```python
main()
```
